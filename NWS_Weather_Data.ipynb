```python
import requests
import pandas as pd
from datetime import datetime
import time
from NWS_Weather_Class import NWSWeatherAPI
```

```python
# califoria city coordinates
california_cities = {
    'Sacramento': (38.5816, -121.4944),
    'Los Angeles': (34.0522, -118.2437),
    'San Francisco': (37.7749, -122.4194),
    'San Diego': (32.7157, -117.1611),
    'Fresno': (36.7378, -119.7871),
    'San Jose': (37.3382, -121.8863),
    'Oakland': (37.8044, -122.2711),
    'Bakersfield': (35.3733, -119.0187)
}

#create instance of class to initialize API
nws = NWSWeatherAPI()

# lets do an example of getting 7-day forecast for Sacramento
lat,lon = california_cities['Sacramento']
sac_forecast_df = nws.get_forecast(lat,lon, hourly = False)
sac_forecast_df
```

```python
# lets try getting an hourly forecast this time for LA
lat,lon = california_cities['Los Angeles']
la_hourly_df = nws.get_forecast(lat, lon, hourly = True)
la_hourly_df
```

```python
# lets now do recent observations for SF 
lat, lon = california_cities['San Francisco']
sf_obs_df = nws.get_observations(lat, lon, limit=24) # can change amount of data points we want to see
sf_obs_df
```

```python
# get weather alerts for CA 
CA_alerts_df = nws.get_alerts(state="CA")
CA_alerts_df
```

```python
# lets get a 7 day forecast for multiple cities in CA

all_forecasts = []
for city, (lat, lon) in list(california_cities.items())[:5]:
    print(f"Fetching {city}.", end=" ")
    multi_city_forecast_df = nws.get_forecast(lat, lon, hourly=False)
    if multi_city_forecast_df is not None:
        multi_city_forecast_df['city'] = city
        all_forecasts.append(multi_city_forecast_df)
    else:
        print("None")
    time.sleep(0.5)  

if all_forecasts:
    multi_city_forecast_df = pd.concat(all_forecasts, ignore_index=True)
    display(multi_city_forecast_df)
else:
    print("No forecasts were retrieved")
```
